```python
vstart = df[df['Variance - BL1 Start Date']<0]
vstartRoad =vstart.loc[(vstart['Area'] != 'ADM')]
VSR = vstartRoad.sort_values(['Variance - BL1 Start Date']).reset_index(drop = True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = "whitegrid")

f,ax = plt.subplots(figsize=(23,17))

sns.set_color_codes("pastel")
ax = sns.barplot(x="Variance - BL1 Start Date", y = "Activity Name", data = VSR)
ax.get_figure().savefig('Activites with Start Variance.png',bbox_inches="tight")
```

```python
VFinish = df[df['Variance - BL1 Finish Date']<0]
VFinishRoad = VFinish.loc[(VFinish['Area'] != 'ADM')]
VFR = VFinishRoad.sort_values(['Variance - BL1 Finish Date']).reset_index(drop = True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "whitegrid")
f,ax = plt.subplots(figsize=(20,17))
sns.set_color_codes("pastel")
ax = sns.barplot(x="Variance - BL1 Finish Date", y = "Activity Name", data = VFR, ax= ax)
ax.get_figure().savefig('Activites with Finish Variance.png',bbox_inches="tight")
```
